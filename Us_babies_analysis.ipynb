```python
# what were the five most popular babies name in 2014 in California
#1- slice out the rows for the year 2014
#2- slice out the rows for the state california
#3- sort the rows in descending order by count
#4- get the first five rows
```

```python
state_babies_2014 = state_babies.loc[state_babies["Year"] == 2014 , :]
```

```python
california_babies = state_babies_2014.loc[state_babies_2014["State"] == 'CA' , :]
```

```python
california_babies
```

```python
sorted_babies = california_babies.sort_values('Count',ascending=False)
```

```python
sorted_babies[0:5]
```

```python
#in california what were the most pupular female and male baby names for each year
#1- slice out rows for california
#2- group resulting dataframe by year and gender
#3- compute the most popular name for each group


### now we use the groupby method , it allows us to group the data in a data frame based on the criteria we specify
ca_babies = state_babies.loc[state_babies['State'] == 'CA' , :]
```

```python
def popular(s):
    return s.iloc[0]
```

```python
ca_gropped = ca_babies.sort_values('Count',ascending = False).groupby(['Year','Gender']).agg(popular)
```
